```
#テキストデータの読み込み・保存先ディレクトリーを作成
author = 'akutagawa' #著者名を指定
title = 'haguruma' #作品名を指定
load_dir = str('/content/drive/My Drive/DSAP/DSAP_Presentation/dataset/'+author+'/'+title+'.txt')
save_dir = str('/content/drive/My Drive/DSAP/DSAP_Presentation/dataset_edited/'+author+'/'+title+'.txt')

#テキストデータの読み込み・保存先のディレクトリーを表示
print(load_dir)
print(save_dir)

#編集が完了した作品を記録
#akutagawa: haguruma, hana, rashomon, torokko, un, tsugaru
#dazai: hanafubuki, hashire_merosu, ningen_shikkaku, shayo, tsugaru
#fukuzawa: gakumonno_dokuritsu, gakumonno_susume, gakusha_anshin_ron, nihon_danshi_ron, onna_daigaku_hyoron
#mori: gan, hyakumonogatari, maihime, moso, seinen
#natsume: bocchan, gubijinso, kokoro, nowaki, wagahaiwa_nekodearu
```

```
with codecs.open(load_dir, 'r', 'shift_jis') as f_r:
    txt = f_r.read()
```

```
# ルビなどの削除
txt = re.sub(u'《[^》]+》', '', txt)
txt = re.sub(u'｜', '', txt)
txt = re.sub(u'［.+?］', '', txt)
txt = re.sub('\u3000', '', txt)

# ヘッダ部分の削除
txt = re.sub(u'-----[\s\S]*-----', '', txt)

# フッタ部分の削除
txt = re.split(u'底本：',txt)[0]

# 改行の削除
#txt = re.sub(u'\n', '', txt)
#txt = re.sub(u'\r', '', txt)
```

```
with open(save_dir, mode='w') as f:
    f.write(txt)
    f.close()
```
